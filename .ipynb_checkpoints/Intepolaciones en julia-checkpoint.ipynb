Mediante el uso de HTTP y JSON solicitamos los datos y analizamos los necesarios, obteniendo el arreglo de las temperaturas en una semana en la ciudad.

```julia
using HTTP
using JSON

location = "Brussels"

start_date = "2024-06-05"  
end_date = "2024-06-11" 

url = "http://api.worldweatheronline.com/premium/v1/past-weather.ashx?q=$(location)&tp=24&date=$(start_date)&enddate=$(end_date)&format=json&key=790e4819ffc841c5b1810355241106"

response = HTTP.request("GET", url)

if response.status == 200
    data = JSON.parse(String(response.body))
    println("Datos de la semana del $start_date al $end_date de $location obtenidos correctamente")
else
    println("Error al obtener datos de la semana del $start_date al $end_date: $(response.status)")
end

temperaturas = []

for day in data["data"]["weather"]
    
    avg_temp = day["avgtempC"]
    push!(temperaturas, parse(Int64,avg_temp))
        
end

println(temperaturas)
```

<h2>Interpolación de Taylor</h2>

Para analizar y predecir estos datos meteorológicos, podemos utilizar la interpolación de Taylor, una técnica matemática que nos permite construir un polinomio que se ajusta a los datos conocidos y nos permite hacer predicciones para valores desconocidos. En este caso, las temperaturas a lo largo de una semana en Bruselas.<br>
<br>
La función de Taylor:

Se utiliza para construir el polinomio de interpolación. Esta función se define como:<br>

$$f(x) = f(a) + f'(a)(x-a) + \frac{f''(a)(x-a)^2}{2!} + \frac{f'''(a)(x-a)^3}{3!} +...$$
<br>
Donde $f(x)$ es la función que se está aproximando, $a$ es el punto de expansión en este caso los dias, $f'(a)$, $f''(a)$, $f'''(a)$,... son las derivadas sucesivas de las temperaturas, y $x$ es el valor que queremos predecir.<br>
<br>
Taylor en codigo de julia:<br><ul>
<li>Factorial(n): calcula el factorial de un número entero n.<br></li>
<li>Derivada(temperaturas): calcula las derivadas sucesivas de una serie de temperaturas.<br></li>
<li>Taylor_series(dias, diferencias, x, grado): calcula el polinomio de Taylor de grado grado para una serie de temperaturas en un punto x, utilizando las derivadas sucesivas calculadas previamente.</li><br></ul>

```julia
using Plots
using SymPy

# Función para calcular el factorial
function factorial(n)
    fact = 1
    for i in 1:n
        fact *= i
    end
    return fact
end

# Función para calcular las derivadas
function derivada(temperaturas)
    diferencias = [temperaturas]
    for i in 1:(length(temperaturas) - 1)
        push!(diferencias, diff(diferencias[end]))
    end
    return diferencias
end

# Función para construir el polinomio de Taylor
function taylor_series(dias, diferencias, x, grado)
    taylor_polinomio = diferencias[1][1]
    for i in 1:grado
        term = diferencias[i+1][1] / factorial(i)
        for j in 0:(i-1)
            term *= (x - dias[j+1])
        end
        taylor_polinomio += term
    end
    return taylor_polinomio
end

# Calcular las diferencias finitas
diferencias = derivada(temperaturas)

# Construir los polinomio de Taylor
dias = 1:length(temperaturas)
x = symbols("x")
taylor_polinomio = taylor_series(dias, diferencias, x, 6)

#Mostrar la grafica
plot(dias, temperaturas, label="Datos originales", marker=:circle)
plot!(taylor_polinomio, label="Interpolación de Taylor", linewidth=2)
xlabel!("Tiempo (Días)")
ylabel!("Temperatura (°C)")
title!("Interpolación de Taylor")
```

<h2>Interpolación de Lagrange</h2>

Para analizar y predecir estos datos meteorologicos, podemos utilizar la interpolación de Lagrange, una técnica matemática que nos permite construir un polinomio que se ajusta a los datos conocidos y nos permite hacer predicciones para valores desconocidos. En este caso las temperaturas a lo largo de una semana en Bruselas.<br>
<br>
La función de Lagrange:<br>
Se utiliza para construir el polinomio de interpolación. Esta función se define como:<br>
<br>
$$L_i(x) = \prod_{j=1, j\neq i}^n \frac{x - x_j}{x_i - x_j}$$<br>
<br>
Donde $x_i$ es el valor de la temperatura en el punto $i$, y $x$ es el valor que queremos predecir.<br>
<br>
La función de interpolación de Lagrange:<br>
Se define como la suma de los productos de cada valor de temperatura $y_i$ con su correspondiente función de Lagrange $L_i(x)$:<br>
<br>
$$P(x) = \sum_{i=1}^n y_i L_i(x)$$<br>
<br>
Donde $y_i$ es el valor de la temperatura en el punto $i$, y $P(x)$ es el polinomio de interpolación que se ajusta a los datos conocidos.<br>
<br>
Lagrange en codigo de julia:<br><ul>
<li>Algoritmo(i, j, x): calcula el factor de Lagrange para un par de índices i y j, y una lista de puntos x.</li>
<li>Factores_lgj: calcula los factores de Lagrange para cada punto de interpolación.</li>
<li>Polinomio: calcula el polinomio de interpolación de Lagrange utilizando los factores de Lagrange y los valores de temperatura.</li></ul>



```julia
using SymPy
using Plots

y = temperaturas
x = 1:length(y)


#Funcion para calcular el factor de Lagrange
function algoritmo(i, j, x)
    xsimb = symbols("x")
    if i != j
        return (xsimb - x[i]) / (x[j] - x[i]) 
    else
        return 1
    end
end

#Funcion para calcular los puntos de interpolacion 
function interpol_lagrange(x, y; puntos_prueba=200)
    xsimb = symbols("x")
    puntos = length(x)
    factores_lgj = []
    for k in 1:puntos
        factor_lg = prod([algoritmo(i, k, x) for i in 1:puntos])
        push!(factores_lgj, factor_lg)
    end
    polinomio = sum(y.* factores_lgj)
    x_prueba = range(minimum(x), stop=maximum(x), length=puntos_prueba)
    y_polinomios = [polinomio(xsimb => i) for i in x_prueba]
    return x_prueba, y_polinomios 
end

#Genera los puntos de Lagrage
x_prueba, y_polinomios = interpol_lagrange(x, y)


#Mostrar la grafica
plot(x_prueba, y_polinomios, label="Interpolación de Lagrange", linewidth=2)
scatter!(x, y, label="Datos originales", marker=:circle)
xlabel!("Tiempo (Días)")
ylabel!("Temperatura (°C)")
title!("Interpolación de Lagrange")
```

<h2>Interpolación de Hermite</h2>

La interpolación de Hermite es una técnica útil para aproximar valores desconocidos a partir de datos conocidos, especialmente cuando hay cambios bruscos en los datos. En el caso del análisis meteorológico, esta técnica puede ser útil para predecir la temperatura en momentos específicos dentro de un rango de tiempo dado.<br>
    
La función de interpolación de Hermite para n puntos de datos con sus respectivas derivadas de primer orden se define como:<br>

$$P(x) = \sum_{i=1}^n H_i(x) y_i + H_i'(x) y_i'$$

Donde:<br>
<ul>
    <li>$P(x)$ es el polinomio de interpolación.</li>
    <li>$H_i(x)$ son las funciones base de Hermite para cada punto i.</li>
    <li>$y_i$ son los valores de temperatura en los puntos i.</li>
    <li>$y_i'$ son las derivadas de primer orden de la temperatura en los puntos i.</li>
</ul>
    
Las funciones base de Hermite se construyen utilizando las siguientes propiedades:<br>

$$H_i(x_j) = \begin{cases} 1 & \text{if } i = j \ 0 & \text{if } i \neq j \end{cases}$$

$$H'_i(x_j) = \begin{cases} 0 & \text{if } i = j \ \text{undefined} & \text{if } i \neq j \end{cases}$$

$$H_i(x) = \begin{cases} 1 & \text{if } x = x_i \ 0 & \text{if } x \neq x_i \end{cases}$$


Una forma común de construir las funciones base de Hermite es:<br>

$$H_i(x) = \prod_{j=1, j\neq i}^n \frac{x - x_j}{x_i - x_j}$$

Para la derivada de primer orden:<br>

$$H_i'(x) = \frac{1}{x_i - x_j} \prod_{j=1, j\neq i}^n (x - x_j)$$
    
Hermite en codigo de julia:<br>
<ul>
<li>Función "calcularDerivadas": esta función calcula las derivadas en los puntos dados.</li>
<li>Función "calcularCoeficientesHermite": esta función calcula los coeficientes de Hermite a partir de las derivadas y los puntos dados.</li>
<li>Función "interpolaciónDeHermite":
Esta función calcula la interpolación de Hermite a partir de los coeficientes de Hermite y el punto a interpolar.</li>

```julia
using Plots

function calcularDerivadas(puntosX, valoresEnPuntosX)
    derivadasEnPuntosX = Float64[] # Inicializa un array vacío para las derivadas

    for i in 1:length(valoresEnPuntosX)-1
        derivada = (valoresEnPuntosX[i+1] - valoresEnPuntosX[i]) / (puntosX[i+1] - puntosX[i])
        push!(derivadasEnPuntosX, derivada) 
    end
    push!(derivadasEnPuntosX, derivadasEnPuntosX[end]) 

    return derivadasEnPuntosX
end

function calcularCoeficientesHermite(puntosX, valoresEnPuntosX, derivadasEnPuntosX)
    # Inicializa los coeficientes de Hermite como flotantes
    coeficientesHermite = Float64.(valoresEnPuntosX)
    
    # Calcula los coeficientes de Hermite
    for i in 2:length(puntosX)
        for j in length(puntosX):-1:i
            numerador = coeficientesHermite[j] - coeficientesHermite[j-1]
            denominador = puntosX[j] - puntosX[j-i+1]
            coeficientesHermite[j] = numerador / denominador
        end
    end
    
    return coeficientesHermite
end

function interpolaciónDeHermite(puntosX, valoresEnPuntosX, puntoAInterpolar)
    derivadasEnPuntosX = calcularDerivadas(puntosX, valoresEnPuntosX)
    coeficientesHermite = calcularCoeficientesHermite(puntosX, valoresEnPuntosX, derivadasEnPuntosX)
    
    # Calcula la interpolación de Hermite
    aproximación = coeficientesHermite[end]
    for i in length(coeficientesHermite)-1:-1:1
        aproximación = aproximación * (puntoAInterpolar - puntosX[i]) + coeficientesHermite[i]
    end
    
    return aproximación
end

días = collect(range(1, stop=length(temperaturas))) 

rangoX = range(1, stop=length(temperaturas), length=100)
rangoY = [interpolaciónDeHermite(días, temperaturas, x) for x in rangoX]
plot(días, temperaturas, seriestype=:scatter, label="Datos Originales", xlabel="Semana (Día)", ylabel="Temperatura (C°)",title="Interpolación de Hermite", legend=:topleft, grid=true)
plot!(rangoX, rangoY, label="Interpolación de Hermite")
```

<h2>Interpolación Polinómica a Trozos</h2>

Para analizar y predecir estos datos meteorologicos, podemos utilizar la interpolación polinómica a trozos, una técnica matemática que nos permite construir un polinomio que se ajusta a los datos conocidos y nos permite hacer predicciones para valores desconocidos. En este caso las temperaturas a lo largo de una semana en Bruselas.<br>

$$S(x) = \begin{cases} S_0(x) & x_0 \leq x \leq x_1 \ S_1(x) & x_1 \leq x \leq x_2 \ \vdots & \vdots \ S_{n-1}(x) & x_{n-1} \leq x \leq x_n \end{cases}$$

Donde $S_0, S_1, \dots, S_{n-1}$ son polinomios de grado menor o igual que $p$ que se definen en cada intervalo $[x_i, x_{i+1}]$.


Función interpolacion_polinomica_por_trozos:<br>
La función itera sobre cada intervalo [x[i], x[i+1]] y calcula los parámetros m y b del polinomio lineal que se utilizará en ese intervalo. Los parámetros se calculan utilizando las condiciones de frontera:

$$m_i = \frac{y_{i+1} - y_i}{x_{i+1} - x_i} \quad \text{y} \quad b_i = y_i - m_i x_i$$

La función devuelve un vector tramos que contiene las tuplas (m, b) para cada intervalo.

Funcion evaluar_interpolacion:

$$S(x) = m_i x + b_i$$

Donde $m_i$ es la pendiente y $b_i$ es el término independiente del polinomio lineal en el intervalo $[x_i, x_{i+1}]$.

La función itera sobre cada punto en puntos, encuentra el intervalo correcto que contiene ese punto, selecciona los parámetros del polinomio correspondiente, evalúa el polinomio en ese punto y almacena el resultado en y_interp.

Polinómica a Trozos en codigo de julia:<br><ul>
<li>tramo: contiene los coeficientes de los polinomios que se utilizarán para la interpolación, calculados por la función.</li>
<li>interpolacion_polinomica_por_trozos(x, y, p): calcula los coeficientes de los polinomios que se utilizarán para la interpolación por trozos siendo x los los dias de la semana, y las temperaturas y p el grado del polinomio.</li>
<li>evaluar_interpolacion(polinomios, x, puntos): calcula la interpolación por trozos de una función desconocida en un conjunto de puntos dados. Donde x son los dias de la semana y puntos los puntos que se desean evaluar en la interpolación.</li><br></ul>

```julia
using Plots

dias_lista = 1:length(temperaturas)

# Crear los tramos para la interpolación
function interpolacion_polinomica_por_trozos(x, y, p)
    tramos = []
    for i in 1:(length(x)-1)
        m = (y[i+1] - y[i]) / (x[i+1] - x[i])
        b = y[i] - m * x[i]
        push!(tramos, (m, b))
    end
    return tramos
end

# Evaluar la interpolación en un conjunto de puntos
function evaluar_interpolacion(polinomios, x, puntos)
    y_interp = []
    for punto in puntos
        for i in 1:(length(x)-1)
            if x[i] <= punto <= x[i+1]
                m, b = polinomios[i]
                push!(y_interp, m * punto + b)
                break
            end
        end
    end
    return y_interp
end

# Aplicar la interpolación polinómica por trozos
polinomios = interpolacion_polinomica_por_trozos(dias_lista, temperaturas, 1)

# Crear un rango de puntos para evaluar la interpolación
puntos = collect(range(1, stop=7, length=100))

# Evaluar la interpolación en los puntos
y_interp = evaluar_interpolacion(polinomios, dias_lista, puntos)

# Gráfica
plot(puntos, y_interp, label="Interpolación Polinómica por Trozos")
scatter!(dias_lista, temperaturas, marker=:circle, label="Temperaturas Promedio")
xlabel!("Semana(Días)")
ylabel!("Temperatura (°C)")
title!("Interpolación Polinómica por Trozos")
```
